```python
import pandas as pd
import numpy as np

tt = pd.read_excel ("Truth_Table.xlsx")
print (tt)
```

```python
p1=tt.loc[tt['X'] == 1]
p1
```

```python
p1.drop(['X'], axis=1,inplace=True)
p1
```

```python
def b2d(n):
    return int(n,2)
```

```python
p2=p1.astype(str)
p2
```

```python
p2['Merged'] = p2.apply(''.join, axis=1)
p2
```

```python
p2['decimal'] = p2.apply(lambda row : b2d(row['Merged']), axis = 1)
p2
```

```python
p3=p2.decimal
p3
```

```python
p4=p3.to_numpy()
print("So,minterms are:",p4[:])
```
